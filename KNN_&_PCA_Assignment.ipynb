**Q1.What is K-Nearest Neighbors (KNN) and how does it work in both
classification and regression problems?**

**A->**

KNN assumes that similar data points exist close to each other in feature space.
To make a prediction for a new (unseen) data point, the algorithm:

- **1**.Looks at the K closest labeled data points (neighbors).

- **2**.Uses those neighbors to predict the output.

KNN relies heavily on distance metrics, commonly:

- Euclidean distance

- Manhattan distance

- Minkowski distance

- Cosine similarity (sometimes)

For classification problems, KNN assigns the class based on majority vote among the K nearest neighbors.

**Q2. What is the Curse of Dimensionality and how does it affect KNN
performance?**

**A->**

When the number of dimensions (features) increases:

**1.Data becomes sparse**

- Points are far apart; density drops.

**2.Distance metrics lose discrimination power**

- Distances between near and far points become almost the same.

**3.Volume of space increases exponentially**

- Requires exponentially more data to represent patterns.


ðŸš§ **How It Affects KNN Performance**

KNN depends on the idea that nearby points have similar values.
But in high dimensions, the concept of â€œnearbyâ€ breaks down.

**1.Distances become less meaningful**

- In high dimensions, the ratio between the nearest and farthest neighbor distances â†’ approaches 1.

- KNN canâ€™t distinguish close vs weakly similar data points.

**2.Increased computational cost**

- KNN must compute distances to all points across all dimensions.

- More dimensions â†’ more calculations â†’ slower prediction.

**3.Overfitting risk**

- High-dimensional space is sparse.

- Finding meaningful local neighborhoods becomes difficult.

- KNN tends to fit noise instead of patterns.

**4.More data required**

- To cover a high-dimensional space, exponentially more training samples are needed.

- Without enough data, KNN becomes unreliable.

**Q3. What is Principal Component Analysis (PCA)? How is it different from
feature selection?**

**A->**

Principal Component Analysis (PCA) is a dimensionality reduction technique used in data analysis and machine learning. It transforms a large set of features into a smaller set while preserving as much variance

âœ… **What is PCA?**

PCA works by:

- 1.Identifying directions (principal components) in which the data varies the most.

- 2.These directions are new axes, created from combinations of your original features.

- 3.The first principal component captures the maximum variance in the data.

- 4.The second captures the next highest variance, and so on â€” all components are uncorrelated.

**So, PCA converts your data into new transformed features like:**

PC1

PC2

PC3

| Aspect                 | **PCA**                                             | **Feature Selection**                            |
| ---------------------- | --------------------------------------------------- | ------------------------------------------------ |
| What it does           | Creates **new features** by combining original ones | Selects **important original features**          |
| Type                   | **Feature extraction**                              | **Feature selection**                            |
| Features after process | Transformed: PC1, PC2â€¦ (you lose original meaning)  | Original features retained (easier to interpret) |
| Goal                   | Maximize **variance** in new components             | Improve accuracy, remove irrelevant features     |
| Interpretability       | Low (PCs are combinations, hard to interpret)       | High (you know which features are selected)      |
| Correlation            | New components are **uncorrelated**                 | Selected features may still be correlated        |



**Q4.What are eigenvalues and eigenvectors in PCA, and why are they
important?**

**A->**

In PCA **(Principal Component Analysis)**, eigenvalues and eigenvectors are the core mathematical pieces that determine how the data is transformed.
They help identify the **principal components,** which are the directions of maximum variance in your data.


âœ… **What are Eigenvectors in PCA?**

An **eigenvector** is a direction in the feature space (like an arrow) along which the data varies the most.

- Each eigenvector represents a principal component direction.

- They show where the variance is happening.

**Example:**
If your data varies mostly along a diagonal line, the eigenvector will point along that line.

âœ… **What are Eigenvalues in PCA?**

An **eigenvalue** tells how much variance lies along its corresponding eigenvector.

- A large eigenvalue â†’ high variance â†’ important component

- A small eigenvalue â†’ low variance â†’ can be removed

Eigenvalue = **importance score** of its eigenvector.


| Concept          | Role                                                                         |
| ---------------- | ---------------------------------------------------------------------------- |
| **Eigenvectors** | Give the **directions** of new axes (principal components)                   |
| **Eigenvalues**  | Tell the **importance** or **variance** captured by each principal component |
| **Together**     | They define the structure of the data in its most informative directions     |


**Q5.How do KNN and PCA complement each other when applied in a single
pipeline?**

**A->**

KNN (K-Nearest Neighbors) and PCA (Principal Component Analysis) work very well together in a machine-learning pipeline because PCA solves many of the problems KNN naturally suffers from.

ðŸ”— **How KNN and PCA Complement Each Other**
1ï¸âƒ£ **PCA reduces dimensionality â†’ improves KNN performance**

KNN becomes slow and inaccurate when the dataset has many features (curse of dimensionality).

PCA reduces high-dimensional data into a smaller number of informative components, so:

- Distance calculations become more meaningful

- Noise and irrelevant features are removed

- KNN becomes faster and more accurate

ðŸ‘‰ PCA helps KNN avoid misleading distances in high dimensions.


2ï¸âƒ£ **PCA removes noise â†’ KNN makes cleaner decisions**

KNN relies entirely on distances between data points.

If raw data contains:

- noisy features

- correlated features

- redundant features

PCA filters out the noise by keeping only components with high variance.

This gives KNN a much cleaner input â†’ better classification or regression.

3ï¸âƒ£ **PCA helps prevent overfitting in KNN**

KNN is a non-parametric, memory-based algorithm and can overfit when:

- Too many features

- Too few samples

PCA reduces features into a compressed representation â†’ lowering the risk of overfitting.

4ï¸âƒ£ **PCA creates uncorrelated (orthogonal) features**

KNN uses Euclidean distance.
Distance calculations become more stable if features are:

- uncorrelated

- scaled similarly

PCA transforms features into orthogonal axes (principal components), making distance-based methods like KNN more reliable.

5ï¸âƒ£ **PCA makes KNN faster (computationally)**

KNN complexity = O(n Ã— d)

- n = number of samples

- d = number of dimensions

By reducing d (features), PCA significantly speeds up KNN.

**Useful for:**

- real-time predictions

- large datasets


| Role           | PCA                         | KNN                           |
| -------------- | --------------------------- | ----------------------------- |
| What it does   | Reduces data, removes noise | Classifies based on neighbors |
| Problem solved | Curse of dimensionality     | Distance-based prediction     |
| Benefit        | Improved speed, accuracy    | Works better with PCA inputs  |



**Q6.Train a KNN Classifier on the Wine dataset with and without feature
scaling. Compare model accuracy in both cases.?**
(Include your Python code and output in the code box below.)


**A->**

âœ… **Accuracy Comparison**

- Without Scaling: 0.74

- With Scaling: 0.96

Scaling significantly improves KNN accuracy because KNN relies on distance measurements, and unscaled features distort those distances.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load dataset
wine = load_wine()
X, y = wine.data, wine.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ---------- WITHOUT SCALING ----------
knn_no_scaling = KNeighborsClassifier(n_neighbors=5)
knn_no_scaling.fit(X_train, y_train)
pred_no_scaling = knn_no_scaling.predict(X_test)
acc_no_scaling = accuracy_score(y_test, pred_no_scaling)

# ---------- WITH SCALING ----------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
pred_scaled = knn_scaled.predict(X_test_scaled)
acc_scaled = accuracy_score(y_test, pred_scaled)

acc_no_scaling, acc_scaled
```

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load dataset
wine = load_wine()
X, y = wine.data, wine.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# -------- ORIGINAL DATA WITH SCALING --------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(X_train_scaled, y_train)
pred_original = knn_original.predict(X_test_scaled)
acc_original = accuracy_score(y_test, pred_original)

# -------- PCA WITH TOP 2 COMPONENTS --------
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
pred_pca = knn_pca.predict(X_test_pca)
acc_pca = accuracy_score(y_test, pred_pca)

acc_original, acc_pca
```

**Q9.Train a KNN Classifier with different distance metrics (euclidean,
manhattan) on the scaled Wine dataset and compare the results.?**
(Include your Python code and output in the code box below.)

**A->**



```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load dataset
X, y = load_wine(return_X_y=True)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scaling
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

results = {}

# Train KNN with different distance metrics
for metric in ["euclidean", "manhattan"]:
    knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
    knn.fit(X_train_s, y_train)
    y_pred = knn.predict(X_test_s)

    acc = accuracy_score(y_test, y_pred)
    clf_report = classification_report(y_test, y_pred)

    results[metric] = (acc, clf_report)

# Display results
for metric, (acc, report) in results.items():
    print("=" * 60)
    print(f"KNN Results using {metric.upper()} distance")
    print(f"Accuracy: {acc:.4f}")
    print(report)
```

ðŸ“Œ Conclusion

Euclidean distance typically performs slightly better on the scaled Wine dataset.

Manhattan distance is still strong, but usually 1â€“2% lower accuracy.

Scaling is essential for KNN â€” without it, results fall significantly.

**Q10.You are working with a high-dimensional gene expression dataset to
classify patients with different types of cance
Due to the large number of features and a small number of samples, traditional models**

Explain how you would:
- â— Use PCA to reduce dimensionality
- â— Decide how many components to keep
- â— Use KNN for classification post-dimensionality reduction
- â— Evaluate the model
- â— Justify this pipeline to your stakeholders as a robust solution for real-world

(Include your Python code and output in the code box below.)


**A->**

âœ… **Explanation of the Pipeline**
1. **Use PCA to reduce dimensionality**

High-dimensional gene expression datasets often contain thousands of genes but fewer than 200 patients.
This leads to:

- Severe overfitting,

- Strong multicollinearity,

- Very noisy features.

**Principal Component Analysis (PCA) helps by:**

- Transforming features into uncorrelated components,

- Retaining directions of maximum variance,

- Dramatically reducing dimensionality while keeping biological signal.

2. **Decide how many PCA components to keep**

Two common strategies:

**A. Retain a cumulative variance threshold**

- Keep enough components to explain 90â€“95% of total varianceâ€”this ensures major biological structure is preserved.

**B. Use cross-validated model performance**

Pick the number of PCA components that yields the best validation accuracy for KNN.

We will combine both:

- 1.Inspect explained variance curve,

- 2.Select number of components with best CV score.

3. **Use KNN for classification after PCA**

**Why KNN after PCA?**

- PCA produces compact Euclidean geometry â‡’ KNN performs better.

- Reduced noise helps KNN avoid curse of dimensionality.

Steps:

- **1**.Scale data using StandardScaler,

- **2**.Apply PCA,

- **3**.Train a KNN classifier (typically k=3,5,7),

- **4**.Evaluate performance.

4. **Evaluate the Model**

We use:

- Stratified train-test split

- Accuracy, precision, recall, F1

- Confusion matrix

(Optional) Cross-validation curves for PCA components

5. **Justify the pipeline to stakeholders**

This pipeline is appropriate for biomedical, high-dimensional, low-sample datasets because:

âœ” **Reduces overfitting**

PCA removes noise and irrelevant gene variation â†’ improves generalization.

âœ” **Reduces computational cost**

Dimensionality reduces from thousands â†’ few dozen components.

âœ” **Increases interpretability**

PCA components can correspond to biological variation patterns (e.g., signaling pathways).

âœ” **Stable and reproducible**

KNN is simple and non-parametric; PCA is deterministic.
This makes the pipeline less prone to instability seen in deep learning models with small datasets.

âœ” **Widely used in genomics**

PCA + KNN is a standard, peer-reviewed approach in cancer genomics and single-cell RNA-seq analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score

# ---------------------------------------------------------
# Simulated high-dimensional cancer gene expression dataset
# (Replace this with real gene expression data when available)
# ---------------------------------------------------------
np.random.seed(42)
n_samples = 150
n_genes = 5000  # High-dimensional
n_classes = 3

X = np.random.randn(n_samples, n_genes)
y = np.random.randint(0, n_classes, n_samples)

# ---------------------------------------------------------
# Train-test split
# ---------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# ---------------------------------------------------------
# Scaling
# ---------------------------------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------------------------------------------------------
# PCA exploration: find number of components
# ---------------------------------------------------------
pca = PCA().fit(X_train_scaled)

explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot explained variance curve
plt.figure(figsize=(8,5))
plt.plot(explained_variance_ratio, marker='o')
plt.axhline(0.95, color='red', linestyle='--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance Curve")
plt.show()

# Choose components that explain 95% variance
n_components_95 = np.argmax(explained_variance_ratio >= 0.95) + 1
print("Components needed for 95% variance:", n_components_95)

# ---------------------------------------------------------
# KNN classification across different PCA dimensions
# ---------------------------------------------------------
component_range = [5, 10, 20, 30, 40, 50, n_components_95]

results = {}
for comp in component_range:
    pca = PCA(n_components=comp)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train_pca, y_train)
    y_pred = knn.predict(X_test_pca)

    acc = accuracy_score(y_test, y_pred)
    results[comp] = acc

print("\nAccuracy by PCA components:")
for comp, acc in results.items():
    print(f"PCA Components: {comp:3d} --> Accuracy: {acc:.4f}")

# ---------------------------------------------------------
# Final Model (Using best number of PCA components)
# ---------------------------------------------------------
best_components = max(results, key=results.get)
print("\nBest PCA components:", best_components)

pca_final = PCA(n_components=best_components)
X_train_pca = pca_final.fit_transform(X_train_scaled)
X_test_pca = pca_final.transform(X_test_scaled)

knn_final = KNeighborsClassifier(n_neighbors=5)
knn_final.fit(X_train_pca, y_train)
y_pred_final = knn_final.predict(X_test_pca)

print("\nFinal Model Performance:")
print(classification_report(y_test, y_pred_final))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_final))
```
